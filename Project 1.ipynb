# Exploratory Data Analysis

### Gender

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

*From above graphs we can see that most of the orders are from Uttar Pradesh, Maharashtra and Karnataka respectively*

### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status', hue='Gender')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

*Occupation*

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

```python
# Product Category
```

```python
Product_cat=df.groupby('Product_Category')['Orders'].sum()
plt.figure(figsize=(12,8))
plt.pie(Product_cat,labels=Product_cat.index,autopct='%1.1f%%', startangle=0, colors=plt.cm.Paired.colors)
plt.title('Product Categories')
plt.axis('equal')
plt.show()
```

*From above graphs we can see that most of the sold products are from Clothing, Food and Electronics*

**Demographic Insights:**

+Married women aged 26-35 years are the most frequent shoppers.
+This age group exhibits a high level of engagement with various products and services

+The highest sales volumes are observed in the states of Uttar Pradesh, Maharashtra, and Karnataka.
